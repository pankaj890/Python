```python
x = np.linspace(-4,4)      # series of values from -4 to 4
```

```python
y1, y2 = x**2, x**3        # x square and x cube
```

```python
#plt.plot(x ,y1)
#plt.plot(x ,y2)
#plt.plot(x, y1, x, y2)
plt.plot(x ,y1,'r--', x, y2,'g')
plt.title("Quadratic and Cubic Powers")
plt.xlabel('X Values -->')
plt.ylabel('Y Values -->')
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2)  # Subplots of subplot
axes[0,0].set_title('Upper Left')
axes[0,1].set_title('Upper Right')
axes[1,0].set_title('Lower Left')
axes[1,1].set_title('Lower Right')

# Below will remove ticks
for ax in axes.flat:
    ax.set_xticks([])
    ax.set_yticks([])
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(9,7))
axes[0,0].set_title('Linear')
axes[0,0].plot(x,x, color='red')
axes[0,1].set_title('Quadratic')
axes[0,1].plot(x,x**2)
axes[1,0].set_title('Cubic')
axes[1,0].plot(x,x**3)
axes[1,1].set_title('4th Power')
axes[1,1].plot(x,x**4)
```

```python
x = np.linspace(-np.pi, np.pi, 200)
```

```python
sine, cosine = np.sin(x), np.cos(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, sine, color = 'red', linestyle = '--', linewidth = 2.5, label = 'Sine Curve')
ax.plot(x, cosine, label = 'Cosine Curve')

ax.legend()
#ax.grid()
ax.set_xticks([-np.pi, -np.pi/2, 0, +np.pi/2, np.pi])
ax.set_yticks(np.arange(-1, 1.2, 0.5))
```

```python
fig, ax = plt.subplots()
ax.plot(x, sine, color = 'red', linestyle = '--', linewidth = 2.5, label = 'Sine Curve')
ax.plot(x, cosine, label = 'Cosine Curve')

ax.legend()
#ax.grid()
ax.set_xticks([-np.pi, -np.pi/2, 0, +np.pi/2, np.pi])
ax.set_yticks([-1, 0, 1])

ax.set_xticklabels(['1st','2nd','3rd','4th','5th'])
ax.set_yticklabels(['A','B','C'], size = 20)

ax.legend(loc = 'upper left')
```
